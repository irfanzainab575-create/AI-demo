```python
class Employee:
    def __init__(self, name, employee_id, salary):
        self.__name = name 
        self.__employee_id = employee_id 
        self.__salary = salary 
        self._department = "General"     
    
    # Getter methods (encapsulation)
    def get_name(self):
        return self.__name
    def get_employee_id(self):
        return self.__employee_id
    def get_salary(self):
        return self.__salary
    def get_department(self):
        return self._department
    
    # Setter methods with validation (encapsulation)
    def set_salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary
            return True
        else:
            print("Salary must be positive!")
            return False
    def set_department(self, new_department):
        self._department = new_department
    
    # Public methods
    def calculate_bonus(self):
        """Calculate bonus (5% of salary)"""
        return self.__salary * 0.05
    
    def display_info(self):
        """Display employee information"""
        print(f"Name: {self.__name}")
        print(f"Employee ID: {self.__employee_id}")
        print(f"Department: {self._department}")
        print(f"Salary: ${self.__salary:,.2f}")
        print(f"Bonus: ${self.calculate_bonus():,.2f}")    

# Create an employee object
emp = Employee("Alice", "E123", 75000)
emp.calculate_bonus()
emp.__name='ali'
emp.get_name()
class Manager(Employee):
    
    def __init__(self, name, employee_id, salary, team_size):
        # Call parent class constructor
        super().__init__(name, employee_id, salary)
        self.__team_size = team_size
        self.set_department("Management") 
    
    # Additional methods specific to Manager
    def get_team_size(self):
        return self.__team_size
    
    def set_team_size(self, new_size):
        if new_size > 0:
            self.__team_size = new_size
            return True
        return False
    
    # Override display method
    def display_info(self):
        super().display_info() 
        print(f"Team Size: {self.__team_size} members")
        print(f"Management Bonus: +${self.__team_size * 500:,.2f}")
m=Manager("Bob", "M456", 90000, 5)
m.display_info()
```

```python
class Human:
    def __init__(self, name, age, gender):
        self.__name = name      
        self.__age = age        
        self._gender = gender   
    
    # Getter methods
    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    def set_age(self, new_age):
        if new_age > 0:
            self.__age = new_age
        else:
            print("Age must be positive!")
    def display_info(self):
        print(f"Name: {self.__name}")
        print(f"Age: {self.__age}")
        print(f"Gender: {self._gender}")
 
# Child class
class Teacher(Human):
    
    def __init__(self, name, age, gender, subject):
        super().__init__(name, age, gender)
        self.__subject = subject   # Private
    
    def display_info(self):
        super().display_info()
        print(f"Subject: {self.__subject}")


# Create object
t1 = Teacher("John", 30, "Male", "Math")
t1.display_info()
```

```python
class Employee:
    def calculate_salary(self):
        pass


class FullTimeEmployee(Employee):
    def __init__(self, salary):
        self.salary = salary

    def calculate_salary(self):
        return self.salary


class PartTimeEmployee(Employee):
    def __init__(self, rate, hours):
        if rate < 0 or hours < 0:
            print("Negative value not allowed")
            self.rate = 0
            self.hours = 0
        else:
            self.rate = rate
            self.hours = hours

    def calculate_salary(self):
        return self.rate * self.hours


class Intern(Employee):
    def __init__(self, stipend):
        self.stipend = stipend

    def calculate_salary(self):
        return self.stipend


# Creating objects
employees = [
    FullTimeEmployee(50000),
    PartTimeEmployee(500, 20),
    Intern(10000)
]

# Polymorphism
for emp in employees:
    print(emp.calculate_salary())
for e in emplys:
    print(e.calculate_salary())
```

```python
class Customer:
    def get_discount(self, amount):
        return amount

class RegularCustomer(Customer):
    def get_discount(self, amount):
        return amount * 0.95

class PremiumCustomer(Customer):
    def get_discount(self, amount):
        return amount * 0.90

class VIPCustomer(Customer):
    def get_discount(self, amount):
        return amount * 0.80

customers = [RegularCustomer(), PremiumCustomer(), VIPCustomer()]
for c in customers:
    print(c.get_discount(5000))
```

```python
class Account:
    def withdraw(self, amount):
        pass

class SavingsAccount(Account):
    def withdraw(self, amount):
        fee = amount * 0.02
        print("Withdraw:", amount - fee)

class CurrentAccount(Account):
    def withdraw(self, amount):
        print("Withdraw:", amount)

class BusinessAccount(Account):
    def withdraw(self, amount):
        fee = amount * 0.05
        print("Withdraw:", amount - fee)

accounts = [SavingsAccount(), CurrentAccount(), BusinessAccount()]
for acc in accounts:
    acc.withdraw(2000)
```
