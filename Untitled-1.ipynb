```python
class Animal: 
    def sound(self):
        print("Animal makes sound")
class Dog(Animal):
    def sound(self):
        print("BHAO BHAO")
class Cat(Animal):
    def sound(self):
        print("Meoww Meowww")
class Cow(Animal):
    def sound(self):
        print("GUAAH GUAAH")
animal = [Dog(), Cat(), Cow()]
for a in animal:
    a.sound()


        
```
