```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")

df=pd.read_csv("D:\pyth1\stock_dataset_dirty.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df['date']=pd.to_datetime(df['Date'])
#df=df.sort_values("Dates")
df.head()
```

```python
df=df.sort_values('Date')
df
```

```python
df.isnull().sum()
```

```python
num_cols=['Open','High','Close','Volume','Market_Cap','Profit_Margin']


for col in num_cols:
    df[col]=df[col].fillna(df[col].mean())
df
```

```python
df['Sector']=df['Sector'].fillna(df['Sector'].mode()[0])
df['Symbol']=df['Symbol'].fillna(df['Symbol'].mode()[0])
df
```

```python
df.isnull().sum()
```

```python
# Average Closing Price by Company
group_symbol = df.groupby("Symbol")["Close"].mean()
print(group_symbol)
```

```python
sector_profit=df.groupby("Sector")["Profit_Margin"].mean()
print(sector_profit)
```

```python
volume_symbol = df.groupby("Symbol")["Volume"].sum()

print(volume_symbol)
```

```python
pivot_table=pd.pivot_table(df,
                           values="High",
                           index="Symbol",
                           columns="Sector",
                           aggfunc="mean")
print(pivot_table)
```

```python
pivot_volume = pd.pivot_table(df,
                              values="Volume",
                              index="Symbol",
                              columns="Sector",
                              aggfunc="sum")

print(pivot_volume)
```

```python
df.groupby("Symbol")["Volume"].sum().sort_values(ascending=True)
```

```python
df.groupby("Sector")["Market_Cap"].mean()
```
