```python
#reshape
#convert 3d into 2d
import numpy as np
arr=np.array([[[1,2,3],[1,2,3],[1,2,3]]])
arr=arr.reshape(3,3)
print(arr)
print(arr.ndim)
#2d array in find silice
arr[1,1]
print(arr.ndim)
print(arr.size)
```
